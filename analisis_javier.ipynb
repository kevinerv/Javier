```python
# Se crean condiciones para filtrar los datos correspondientes a Outer Membrane
cond1 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=1.00)'
cond2 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.99)'
cond3 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.98)'
cond4 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.95)'
cond5 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.90)'
cond5 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.90)'
cond6 = dfDatos['Localization(Probability)'] == 'Outer Membrane                                  (Prob.=0.83)'

# filtrar por varias condiciones con "or" y se obtiene tabla final para Outer Membrane con 164 rows × 12 columns
outer_membrane = dfDatos[cond1 | cond2 | cond3 | cond4 | cond5 | cond6]
outer_membrane
```
