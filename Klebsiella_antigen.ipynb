### Tabla Outer Membrane

```python
# Se crean condiciones para filtrar los datos correspondientes a Outer Membrane
cond1 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=1.00)'
cond2 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=0.99)'
cond3 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=0.98)'
cond4 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=0.95)'
cond5 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=0.90)'
cond6 = dfDatos['Localization_(Probability)'] == 'Outer Membrane                                  (Prob.=0.83)'
cond7 = dfDatos['Localization_(Probability)'] == 'Extracellular                                  (Prob.=0.79)'
cond8 = dfDatos['Localization_(Probability)'] == 'Extracellular                                  (Prob.=0.95)'
cond9 = dfDatos['Localization_(Probability)'] == 'Extracellular                                  (Prob.=0.96)'
cond10 = dfDatos['Localization_(Probability)'] == 'Extracellular                                  (Prob.=0.97)'
cond11 = dfDatos['Localization_(Probability)'] == 'Extracellular                                  (Prob.=1.00)'


# filtrar por varias condiciones con "or" y se obtiene tabla final para Outer Membrane con 164 rows √ó 12 columns
outer_membrane = dfDatos[cond1 | cond2 | cond3 | cond4 | cond5 | cond6 | cond7 | cond8 | cond9 | cond10 | cond11]
outer_membrane = outer_membrane.reset_index()
outer_membrane.index = outer_membrane.index+1
outer_membrane

```

```python
# Se excluyen registros con datos "Yes" en los siguientes campos
# Similar Human Protein,  Similar Mouse Protein, Similar Pig Protein
outer_membrane = outer_membrane[outer_membrane["Similar_Human_Protein"] != "Yes"]
outer_membrane = outer_membrane[outer_membrane["Similar_Mouse_Protein"] != "Yes"]
outer_membrane = outer_membrane[outer_membrane["Similar_Pig_Protein"] != "Yes"]
outer_membrane
```

```python
# Vaxign‚ÄëML Score en orden descendente
outer_membrane = outer_membrane.sort_values(by=['Vaxign_ML_Score'], ascending=False).reset_index()
outer_membrane = outer_membrane.drop(columns=['level_0','index'])
outer_membrane
```

```python
#Exportar tabla a excel

outer_membrane.to_excel("datos_1.xlsx", index=False)
```
